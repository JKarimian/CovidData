### Top Cases

```python
from requests import get
from uk_covid19 import Cov19API
import pandas as pd
from datetime import datetime
pd.options.plotting.backend = "plotly"
```

```python
region = [
    'areaType=ltla',
]
cases_and_deaths = {
    "date": "date",
    "areaName": "areaName",
    "areaCode": "areaCode",
    "newCasesBySpecimenDate": "newCasesBySpecimenDate",
    "newDeathsByDeathDate": "newDeathsByDeathDate",
    "newTestsByPublishDate": "newTestsByPublishDate"
}
api = Cov19API(filters=region, structure=cases_and_deaths)
data = api.get_json()
'Last updated: ' + data['lastUpdate']
```

```python
df = pd.DataFrame(data['data'])
len(df['areaName'].unique())
```

```python
df.info()
```

I want to return a row entry for the largest CaseBySpecimentData

```python
df.head()
```

```python
df[df['date'] == '2020-10-04']['newCasesBySpecimenDate'].argmax()
```

```python
top_places = df[df['date'] == '2020-10-01'].nlargest(7, 'newCasesBySpecimenDate')
top_places = df[df['areaName'].isin(top_places['areaName'])]
top_places.info()
```

```python
pv = top_places.pivot_table(index='date', values='newCasesBySpecimenDate', columns='areaName')
pv.plot(title=" New Cases by Specimen Data", labels={'value':'Cases', 'areaName':'Area Name', 'date':'Date'})
```
